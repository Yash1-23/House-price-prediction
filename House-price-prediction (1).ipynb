```python
# Remove the 'sqft_above' column from both X_train and X_test if it was dropped
X_train = X_train.drop(columns=['sqft_above'], errors='ignore')
X_test = X_test.drop(columns=['sqft_above'], errors='ignore')

# Check the columns to ensure they match
print("Training columns:", X_train.columns)
print("Testing columns:", X_test.columns)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

# Initialize and fit the model
clf = GradientBoostingRegressor(n_estimators=100, loss='squared_error')
clf.fit(X_train, y_train)

# Initialize an array to store MSE values
t_sc = np.zeros(clf.n_estimators, dtype=np.float64)

# Calculate MSE for staged predictions
for i, y_pred in enumerate(clf.staged_predict(X_test)):
    t_sc[i] = mean_squared_error(y_test, y_pred)

print(t_sc)
```

```python

testsc  = np.arange((params['n_estimators']))+1
   
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_, 'b-',label = 'set dev train')
plt.plot(testsc,t_sc, 'r-',label = 'set dev test')
```
